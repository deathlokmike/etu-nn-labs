```python
import torchvision
import torch
import matplotlib.pyplot as plt
from collections import Counter

transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])

train_set = torchvision.datasets.MNIST(
    root="./data", train=True, download=True, transform=transform
)

test_set = torchvision.datasets.MNIST(
    root="./data", train=False, download=True, transform=transform
)
train_classes = [label for _, label in train_set]

print(len(train_set))
print(len(test_set))

def pie_plot(_df):
    labels = list(_df.keys())
    values = list(_df.values())
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.grid(False)

    plt.pie(values, labels=labels, autopct="%1.1f%%")
    my_circle = plt.Circle((0, 0), 0.7, color="#ffffff")
    plt.title("Cоотношение количества объектов разных классов")
    plt.gcf().gca().add_artist(my_circle)

    plt.show()


pie_plot(dict(Counter(train_classes)))
```

```python
train_set = torchvision.datasets.PCAM(
    root="./data", split="train", download=True, transform=transform
)

test_set = torchvision.datasets.PCAM(
    root="./data", split="test", download=True, transform=transform
)

train_classes = [label for _, label in train_set]

print(len(train_set))
print(len(test_set))

pie_plot(dict(Counter(train_classes)))
```
